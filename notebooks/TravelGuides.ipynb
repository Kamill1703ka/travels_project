```python
from transformers import pipeline
```

```python
classifier = pipeline(
    "zero-shot-classification",
    model="facebook/bart-large-mnli"
)

print("Модель успешно загружена")
```

```python
labels = [
    "weather request",
    "country information",
    "travel advice",
    "greeting",
    "general question"
]
```

```python
text = "Что посмотреть в Италии?"

result = classifier(text, labels)

print(result)
```

```python
def classify_message(text):

    result = classifier(text, labels)

    return result["labels"][0]
```

```python
sentiment_labels = [
    "positive sentiment",
    "negative sentiment",
    "neutral sentiment"
]
```

```python
def detect_sentiment(text):

    result = classifier(text, sentiment_labels)

    return result["labels"][0]
```

```python
from transformers import pipeline

# Модель для тональности
sentiment_analyzer = pipeline(
    "sentiment-analysis",
    model="nlptown/bert-base-multilingual-uncased-sentiment"
)

texts = [
    "Мне очень понравилась Италия!",
    "Этот бот ужасный",
    "Сегодня обычная погода"
]

for t in texts:
    result = sentiment_analyzer(t)
    print("Текст:", t)
    print("Тональность:", result[0]['label'])
    print("-------------------")
```

```python
def sentiment_label_to_simple(star_label):
    if star_label in ["1 star", "2 stars"]:
        return "negative"
    elif star_label in ["3 stars"]:
        return "neutral"
    else:
        return "positive"

for t in texts:
    raw_result = sentiment_analyzer(t)
    sentiment = sentiment_label_to_simple(raw_result[0]['label'])
    print("Текст:", t)
    print("Тональность:", sentiment)
```
